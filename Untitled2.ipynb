```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic voltage data for 3 phases
np.random.seed(0)  # For reproducibility
n_hours = 24 * 5  # One week
timestamps = pd.date_range(start='2022-01-01', periods=n_hours, freq='H')
voltages_phase_1 = np.random.normal(230, 5, n_hours)  # Normal distribution around 230V
voltages_phase_2 = np.random.normal(230, 5, n_hours)
voltages_phase_3 = np.random.normal(230, 5, n_hours)

# Introduce random outliers
outlier_indices = np.random.choice(n_hours, size=int(n_hours * 0.01), replace=False)  # Assuming 5% outliers
voltages_phase_1[outlier_indices] += np.random.choice([-50, 50], size=len(outlier_indices))  # Add outliers to phase 1
voltages_phase_2[outlier_indices] += np.random.choice([-50, 50], size=len(outlier_indices))  # Add outliers to phase 2
voltages_phase_3[outlier_indices] += np.random.choice([-50, 50], size=len(outlier_indices))  # Add outliers to phase 3

# Create DataFrame for voltage data
voltage_data = {
    'timestamp': timestamps,
    'phase_1': voltages_phase_1,
    'phase_2': voltages_phase_2,
    'phase_3': voltages_phase_3
}
voltage_df = pd.DataFrame(voltage_data)

# Plot lineplot of voltages of 3 phases
plt.figure(figsize=(12, 6))
plt.plot(voltage_df['timestamp'], voltage_df['phase_1'], label='Phase 1')
plt.plot(voltage_df['timestamp'], voltage_df['phase_2'], label='Phase 2')
plt.plot(voltage_df['timestamp'], voltage_df['phase_3'], label='Phase 3')

# Highlight outliers
plt.scatter(voltage_df.loc[outlier_indices, 'timestamp'], voltage_df.loc[outlier_indices, 'phase_1'], color='red', label='Outliers')
plt.scatter(voltage_df.loc[outlier_indices, 'timestamp'], voltage_df.loc[outlier_indices, 'phase_2'], color='red')
plt.scatter(voltage_df.loc[outlier_indices, 'timestamp'], voltage_df.loc[outlier_indices, 'phase_3'], color='red')

# Add labels and legend
plt.title('Voltage Readings of 3 Phases with Outliers')
plt.xlabel('Timestamp (Hourly)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.xticks(ha="right")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)  # For reproducibility

# Generate synthetic timestamp data for one year
timestamps = pd.date_range(start='2022-01-01', end='2022-12-31', freq='D')

# Generate synthetic outlier counts for each month
outlier_counts = {
    'timestamp': timestamps,
    'is_outlier': np.random.randint(0, 2, len(timestamps))  # Random counts of outliers (0 to 10) for each day
}

# Convert data to DataFrame
outlier_counts_df = pd.DataFrame(outlier_counts)

# Extract month from timestamp
outlier_counts_df['month'] = outlier_counts_df['timestamp'].dt.month

# Group by month and calculate percentage of outliers
monthly_outlier_percentage = outlier_counts_df.groupby('month')['is_outlier'].mean() * 100/10

# Plot count plot
plt.figure(figsize=(15, 5))
sns.barplot(data=monthly_outlier_percentage.reset_index(), x='month', y='is_outlier', color='orange')
plt.title('Monthly Outlier Percentage')
plt.xlabel('Month')
plt.ylabel('Outlier Percentage (%)')
plt.xticks(ha="right")

# Add percentage values on top of each bar
for index, value in enumerate(monthly_outlier_percentage):
    plt.text(index, value, f"{value:.2f}%", color='black', ha="center", va="bottom")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for KW and kVAR
np.random.seed(0)
n_hours = 24 * 7  # One week
timestamps = pd.date_range(start='2024-04-01', periods=n_hours, freq='H')
# timestamps = pd.date_range(start='2024-04-01', end='2024-04-02', freq='15min')
kw_data = np.random.normal(100, 10, n_hours)  # KW data with normal distribution
kvar_data = np.random.normal(50, 5, n_hours)  # kVAR data with normal distribution

# Introduce outliers
outlier_indices = np.random.choice(n_hours, size=int(n_hours * 0.05), replace=False)  # 5% outliers
kw_data[outlier_indices] += np.random.choice([-20, 20], size=len(outlier_indices))  # Add outliers to KW data
kvar_data[outlier_indices] += np.random.choice([-10, 10], size=len(outlier_indices))  # Add outliers to kVAR data

# Create DataFrame
data = {
    'timestamp': timestamps,
    'KW': kw_data,
    'kVAR': kvar_data
}
df = pd.DataFrame(data)

# Plot line plot with outliers
plt.figure(figsize=(8,4))
plt.plot(df['timestamp'], df['KW'], label='KW')
plt.plot(df['timestamp'], df['kVAR'], label='kVAR')
plt.scatter(df.loc[outlier_indices, 'timestamp'], df.loc[outlier_indices, 'KW'], color='red', label='KW Outliers')
plt.scatter(df.loc[outlier_indices, 'timestamp'], df.loc[outlier_indices, 'kVAR'], color='blue', label='kVAR Outliers')
# plt.title('KW and kVAR with Outliers')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.legend()
plt.xticks(ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data generation
np.random.seed(0)
timestamps = pd.date_range(start='2024-04-01', end='2024-04-02', freq='15min')
num_intervals = len(timestamps)
phase_1_voltage = np.random.normal(230, 5, num_intervals)
phase_2_voltage = np.random.normal(230, 5, num_intervals)
phase_3_voltage = np.random.normal(230, 5, num_intervals)

# Define upper and lower limits for each phase
upper_limit = 250
lower_limit = 210

# Introduce some outliers
phase_1_voltage[20] = 500
phase_2_voltage[35] = -120
phase_2_voltage[50] = -240
# phase_3_voltage[50] = 300

# Create DataFrame
data = pd.DataFrame({
    'Timestamp': timestamps,
    'Phase 1 Voltage (V)': phase_1_voltage,
    'Phase 2 Voltage (V)': phase_2_voltage,
    'Phase 3 Voltage (V)': phase_3_voltage
})

# Line plot
plt.figure(figsize=(8, 4))
sns.lineplot(x='Timestamp', y='Phase 1 Voltage (V)', data=data, label='Phase 1 Voltage')
sns.lineplot(x='Timestamp', y='Phase 2 Voltage (V)', data=data, label='Phase 2 Voltage')
sns.lineplot(x='Timestamp', y='Phase 3 Voltage (V)', data=data, label='Phase 3 Voltage')

# Scatter plot for outliers - Phase 1
phase_1_outliers = data[(data['Phase 1 Voltage (V)'] > upper_limit) | (data['Phase 1 Voltage (V)'] < lower_limit)]
plt.scatter(phase_1_outliers['Timestamp'], phase_1_outliers['Phase 1 Voltage (V)'], color='red', marker='x', label='Outliers')

# Scatter plot for outliers - Phase 2
phase_2_outliers = data[(data['Phase 2 Voltage (V)'] > upper_limit) | (data['Phase 2 Voltage (V)'] < lower_limit)]
plt.scatter(phase_2_outliers['Timestamp'], phase_2_outliers['Phase 2 Voltage (V)'], color='red', marker='x')

# Scatter plot for outliers - Phase 3
phase_3_outliers = data[(data['Phase 3 Voltage (V)'] > upper_limit) | (data['Phase 3 Voltage (V)'] < lower_limit)]
plt.scatter(phase_3_outliers['Timestamp'], phase_3_outliers['Phase 3 Voltage (V)'], color='red', marker='x')

# plt.title('AMI Meter Voltage Readings')
plt.xlabel('Timestamp')
plt.ylabel('Voltage (V)')
plt.legend()
plt.xticks(ha='right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Function to generate synthetic voltage data for each phase
def generate_voltage_data(num_samples, event_type):
    # Generating synthetic timestamps for a day
    timestamps = pd.date_range(start='2022-01-01', periods=num_samples, freq='15T')
    
    # Generating synthetic voltage data for each phase based on event type
    if event_type == "Voltage Fluctuations":
        # Simulating random voltage fluctuations for a two-hour period
        voltage_data_A = np.zeros(num_samples) + 120  # Initialize voltage data with normal level
        voltage_data_B = np.zeros(num_samples) + 180
        voltage_data_C = np.zeros(num_samples) + 277
        
        # Introduce voltage fluctuations for a two-hour window (randomly chosen)
        start_index = np.random.randint(0, num_samples - 8)  # Start index for fluctuations (random)
        voltage_data_A[start_index:start_index+16] += np.random.normal(0, 30, 16)  # Fluctuations in voltage
        voltage_data_B[start_index:start_index+16] += np.random.normal(0, 10, 16)
#         voltage_data_C[start_index:start_index+8] += np.random.normal(0, 5, 8)
        
        return timestamps, voltage_data_A, voltage_data_B, voltage_data_C

# Function to plot voltage data with timestamps
def plot_voltage_data(timestamps, voltage_data_A, voltage_data_B, voltage_data_C):
    plt.figure(figsize=(10, 4))
    plt.plot(timestamps, voltage_data_A, label='Phase A')
    plt.plot(timestamps, voltage_data_B, label='Phase B')
    plt.plot(timestamps, voltage_data_C, label='Phase C')
    plt.xlabel('Timestamp')
    plt.ylabel('Voltage')
    plt.title('Voltage Fluctuations')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()

# Generate synthetic voltage data for Voltage Fluctuations
num_samples = 96  # Number of samples (15-minute intervals for a day)
timestamps, voltage_data_A, voltage_data_B, voltage_data_C = generate_voltage_data(num_samples, "Voltage Fluctuations")

# Plot voltage data with timestamps for the entire 24-hour period
plot_voltage_data(timestamps, voltage_data_A, voltage_data_B, voltage_data_C)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Function to generate synthetic voltage and current data with a fault
def generate_voltage_current_data_with_fault(num_samples):
    # Generating synthetic timestamps for a day
    timestamps = pd.date_range(start='2022-01-01', periods=num_samples, freq='15T')
    
    # Generating synthetic voltage and current data
    voltage_data = np.zeros(num_samples) + 220  # Initialize voltage data with normal level
    current_data = np.zeros(num_samples) + 10   # Initialize current data with normal level
    
    # Introduce a fault event randomly within a two-hour window
    fault_start_index = np.random.randint(0, num_samples - 8)  # Start index for fault (random)
    fault_duration = np.random.randint(1, 5)  # Duration of fault (random)
    voltage_data[fault_start_index:fault_start_index+fault_duration] -= 50  # Simulate voltage drop due to fault
    current_data[fault_start_index:fault_start_index+fault_duration] += 50   # Simulate increase in current due to fault
    
    return timestamps, voltage_data, current_data, fault_start_index

# Function to plot voltage and current data with timestamps
def plot_voltage_current_data_with_fault(timestamps, voltage_data, current_data, fault_start_index):
    plt.figure(figsize=(10, 4))
    plt.plot(timestamps, voltage_data, label='Voltage')
    plt.plot(timestamps, current_data, label='Current')
    plt.axvline(x=timestamps[fault_start_index], color='r', linestyle='--', label='Fault Start')  # Mark fault start
    plt.xlabel('Timestamp')
    plt.ylabel('Value')
    plt.title('Voltage and Current Data with Fault')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()

# Generate synthetic voltage and current data with fault
num_samples = 96  # Number of samples (15-minute intervals for a day)
timestamps, voltage_data, current_data, fault_start_index = generate_voltage_current_data_with_fault(num_samples)

# Plot voltage and current data with fault
plot_voltage_current_data_with_fault(timestamps, voltage_data, current_data, fault_start_index)
```
